### **1. Adequação do nível (IAN)**

Qual é o perfil geral de defasagem dos alunos (IAN) e como ele evolui ao longo do ano?

(Ex.: quantos alunos estão moderadamente ou severamente defasados?)

---

### Encontrar padrões de dados para alunos com status de defasagem por ano e gerar um indicador para visualiação desse índice por ano;

IAN = Fase efetiva - Fase ideal:

 - Se IAN >= 0 : **Em fase**;

 - Se IAN < 0 e >= -2: **Moderada**;

 - Se IAN < -2 : **Severa**;

### 2. Desempenho acadêmico (IDA)

O desempenho acadêmico médio (IDA) está melhorando, estagnado ou caindo ao longo das fases e anos?

---

### Análise geral do IDA por ano;

 - IDA = (Nota Matemática + Nota Português + Nota Inglês) / 3

*OBS: Dados necessários: Notas (internas da associação);*

### 3. Engajamento nas atividades (IEG)

O grau de engajamento dos alunos (IEG) tem relação direta com seus indicadores de desempenho (IDA) e do ponto de virada (IPV)?

---


### Correlação do IEG com o IDA e IPV;

 - IEG = Σ das pontuações das tarefas realizadas e registradas / Número de tarefas
 *Exemplos: Participação em tarefas de casa, Atividades acadêmicas, Voluntariado;*

 - IDA = (Nota Matemática + Nota Português + Nota Inglês) / 3

 - IPV = Análises longitudinais de progresso acadêmico, engajamento e desenvolvimento emocional;

### 4. Autoavaliação (IAA)

As percepções dos alunos sobre si mesmos (IAA) são coerentes com seu desempenho real (IDA) e engajamento (IEG)?

---

### Correlação do IAA com o IDA e IEG;

 - IAA = Σ(somatória) das pontuações das respostas do estudante / Número total de perguntas

*OBS: Perguntas de autoavaliação de 0 a 10*

*OBS: Avaliações feitas por psicólogos (comportamental, emocional, social);*
